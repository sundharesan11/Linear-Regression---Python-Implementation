```python
import pandas as pd
import numpy as np
```

```python
def calculate_mean(data):
    return sum(data) / len(data)

def calculate_variance(data, mean):
    return sum((x - mean) ** 2 for x in data)

def calculate_covariance(X, mean_x, y, mean_y):
    return sum((X[i] - mean_x) * (y[i] - mean_y) for i in range(len(X)))

def simple_linear_regression(X, y):
    # Step 1: Calculate the mean of X and y
    mean_x = calculate_mean(X)
    mean_y = calculate_mean(y)

    # Step 2: Calculate the covariance of X and y
    cov_xy = calculate_covariance(X, mean_x, y, mean_y)

    # Step 3: Calculate the variance of X
    var_x = calculate_variance(X, mean_x)

    # Step 4: Compute the slope (beta1)
    beta1 = cov_xy / var_x

    # Step 5: Compute the intercept (beta0)
    beta0 = mean_y - beta1 * mean_x

    return beta0, beta1

```

```python
df = pd.read_csv("Advertising.csv")
df.head()
```

```python
X1 = df['TV']
X2 = df['Newspaper']
y = df['Sales']
```

```python
import numpy as np
import plotly.graph_objects as go

def multiple_linear_regression(X1, X2, y):
   
    X = np.column_stack((np.ones_like(X1), X1, X2))
    
    XtX = np.dot(X.T, X)
    
    XtX_inv = np.linalg.inv(XtX)
    
    Xty = np.dot(X.T, y)
    
    beta = np.dot(XtX_inv, Xty)
    
    return beta



coefficients = multiple_linear_regression(X1, X2, y)
print("Coefficients:", coefficients)

X1_mesh, X2_mesh = np.meshgrid(np.linspace(min(X1), max(X1), 100), np.linspace(min(X2), max(X2), 100))
X0_mesh = np.ones_like(X1_mesh)
y_mesh = coefficients[0] + coefficients[1] * X1_mesh + coefficients[2] * X2_mesh

fig = go.Figure()

fig.add_trace(go.Scatter3d(x=X1, y=X2, z=y, mode='markers', marker=dict(size=5), name='Data Points'))

fig.add_trace(go.Surface(x=X1_mesh, y=X2_mesh, z=y_mesh, name='Regression Plane'))

fig.update_layout(scene=dict(xaxis_title='X1', yaxis_title='X2', zaxis_title='y'), margin=dict(l=0, r=0, b=0, t=40))

fig.show()
```

```python
import numpy as np

def initialize_parameters():
    # Initialize parameters randomly
    m = 0
    c = 0
    return m, c

def compute_cost(X, y, m, c):
    # Compute the mean squared error (MSE) cost
    n = len(y)
    predictions = m * X + c
    cost = np.sum((predictions - y) ** 2) / (2 * n)
    return cost

def compute_gradients(X, y, m, c):
    # Compute the gradients of the cost function with respect to parameters m and c
    n = len(y)
    dm = np.sum(-X * (y - (m*X + c))) / n
    dc = np.sum(-(y - (m*X + c))) / n
    return dm, dc

def update_parameters(m, c, dm, dc, learning_rate):
    # Update parameters using gradient descent
    m -= learning_rate * dm
    c -= learning_rate * dc
    return m, c

def train(X, y, learning_rate, iterations):
    # Train the linear regression model using gradient descent
    m, c = initialize_parameters()
    for i in range(iterations):
        # Compute cost
        cost = compute_cost(X, y, m, c)
        # Compute gradients
        dm, dc = compute_gradients(X, y, m, c)
        # Update parameters
        m, c = update_parameters(m, c, dm, dc, learning_rate)
        # Print cost every 100 iterations
        if i % 100 == 0:
            print(f"Iteration {i}: Cost = {cost}")
    return m, c

# Example usage:
# Generate some random data
np.random.seed(0)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

# Define hyperparameters
learning_rate = 0.01
iterations = 1000

# Train the model
m, c = train(X, y, learning_rate, iterations)

print("Trained parameters:")
print("Slope (m):", m)
print("Intercept (c):", c)
```

```python
import matplotlib.pyplot as plt

# Plot the original data
plt.scatter(X, y, color='blue', label='Original data')

# Predict y values using the trained parameters
y_pred = m * X + c

# Plot the regression line
plt.plot(X, y_pred, color='red', label='Regression line')

plt.xlabel('Size (Square feet)')
plt.ylabel('Price ($)')
plt.title('Linear Regression')
plt.legend()
plt.show()
print(m, c)
```

```python
import numpy as np

def initialize_parameters(num_features):
    # Initialize parameters randomly
    return np.zeros((num_features, 1))

def compute_cost(X, y, theta):
    # Compute the mean squared error (MSE) cost
    n = len(y)
    predictions = np.dot(X, theta)
    cost = np.sum((predictions - y) ** 2) / (2 * n)
    return cost


def compute_gradients(X, y, theta):
    # Compute the gradients of the cost function with respect to parameters theta
    n = len(y)
    gradients = np.dot(X.T, (np.dot(X, theta) - y.reshape(-1, 1))) / n
    return gradients


def update_parameters(theta, gradients, learning_rate):
    # Update parameters using gradient descent
    theta -= learning_rate * gradients
    return theta

def train(X, y, learning_rate, iterations):
    # Train the multiple linear regression model using gradient descent
    num_features = X.shape[1]
    theta = initialize_parameters(num_features)
    for i in range(iterations):
        # Compute cost
        cost = compute_cost(X, y, theta)
        # Compute gradients
        gradients = compute_gradients(X, y, theta)
        # Update parameters
        theta = update_parameters(theta, gradients, learning_rate)
        # Print cost every 100 iterations
        if i % 100 == 0:
            print(f"Iteration {i}: Cost = {cost}")
    return theta

# Example usage:
# Generate some random data
np.random.seed(0)
X = 2 * np.random.rand(100, 3)  # 3 features
y = 4 + np.dot(X, np.array([3, 5, 2])) + np.random.randn(100)

# Add bias term (intercept)
X = np.c_[np.ones((X.shape[0], 1)), X]

# Define hyperparameters
learning_rate = 0.01
iterations = 1000

# Train the model
theta = train(X, y, learning_rate, iterations)

print("Trained parameters:")
print(theta)
```

```python
import plotly.graph_objs as go

def plot_regression_line(X, y, theta):
    # Generate points for regression plane
    x1 = np.linspace(min(X[:, 1]), max(X[:, 1]), 10)
    x2 = np.linspace(min(X[:, 2]), max(X[:, 2]), 10)
    x1, x2 = np.meshgrid(x1, x2)
    y_pred = theta[0] + theta[1] * x1 + theta[2] * x2
    
    # Create trace for original data
    trace1 = go.Scatter3d(
        x=X[:, 1],
        y=X[:, 2],
        z=y.flatten(),
        mode='markers',
        marker=dict(
            color='blue',
            size=5,
            opacity=0.8
        ),
        name='Original data'
    )

    # Create trace for regression plane
    trace2 = go.Surface(
        x=x1,
        y=x2,
        z=y_pred,
        opacity=0.6,
        colorscale='Viridis',
        name='Regression plane'
    )

    # Define layout
    layout = go.Layout(
        title='Multiple Linear Regression',
        scene=dict(
            xaxis=dict(title='Feature 1'),
            yaxis=dict(title='Feature 2'),
            zaxis=dict(title='Target')
        )
    )

    # Create figure and plot
    fig = go.Figure(data=[trace1, trace2], layout=layout)
    fig.show()

# Example usage:
# Assuming you have trained the model and obtained theta
plot_regression_line(X, y, theta)
```

```python
import plotly.graph_objs as go

def plot_regression_line(X, y, theta):
    # Generate points for regression plane
    x1_values = np.linspace(min(X[:, 1]), max(X[:, 1]), 10)
    x2_values = np.linspace(min(X[:, 2]), max(X[:, 2]), 10)
    x1_mesh, x2_mesh = np.meshgrid(x1_values, x2_values)
    x0_mesh = np.ones_like(x1_mesh)
    x_flat = np.c_[x0_mesh.flatten(), x1_mesh.flatten(), x2_mesh.flatten()]
    y_pred_mesh = np.dot(x_flat, theta.flatten()).reshape(x1_mesh.shape)
    
    # Create trace for original data
    trace1 = go.Scatter3d(
        x=X[:, 1],
        y=X[:, 2],
        z=y.flatten(),
        mode='markers',
        marker=dict(
            color='blue',
            size=5,
            opacity=0.8
        ),
        name='Original data'
    )

    # Create trace for regression plane
    trace2 = go.Surface(
        x=x1_mesh,
        y=x2_mesh,
        z=y_pred_mesh,
        opacity=0.6,
        colorscale='Viridis',
        name='Regression plane'
    )

    # Define layout
    layout = go.Layout(
        title='Multiple Linear Regression',
        scene=dict(
            xaxis=dict(title='Feature 1'),
            yaxis=dict(title='Feature 2'),
            zaxis=dict(title='Target')
        )
    )

    # Create figure and plot
    fig = go.Figure(data=[trace1, trace2], layout=layout)
    fig.show()

# Example usage:
# Assuming you have trained the model and obtained theta
plot_regression_line(X, y, theta)
```
